```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('80QS.csv')
```

```python
df.head()
```

```python
def is_numeric(col):
    return col.dtype in ['int64', 'float64']

def describe(col):
    if is_numeric(col):
        return {
            'type': 'numeric',
            'dtype': str(col.dtype),
            'min': float(col.min()),
            'max': float(col.max()),
            'mean': float(col.mean()),
            'median': float(col.median()),
            'std': float(col.std()),
        }
    else:
        unqs = col.unique()
        if len(unqs) < 10:
            # Convert unique values to list and handle any non-serializable types
            unique_list = [str(x) if not isinstance(x, (int, float, str, bool)) else x for x in unqs]
            # Convert value_counts to a dictionary
            value_counts_dict = col.value_counts().to_dict()
            # Ensure keys are strings for JSON compatibility
            value_counts_dict = {str(k): float(v) for k, v in value_counts_dict.items()}
            
            return {
                'type': 'categorical',
                'dtype': str(col.dtype),
                'unique_values': unique_list,
                'value_counts': value_counts_dict
            }
        else:
            # For columns with many unique values, just return a sample
            sample_list = [str(x) if not isinstance(x, (int, float, str, bool)) else x for x in col.sample(10)]
            return {
                'type': 'text',
                'dtype': str(col.dtype),
                'sample_values': sample_list,
                'total_unique': len(unqs)
            }

describe(df['AGE'])  
describe(df['Gender'])
        
```

```python
dic_output = {}

for col in df.columns:
    dic_output[col] = describe(df[col])

dic_output
```

```python
import json
with open('describe.json', 'w') as f:
    json.dump(dic_output, f)
```
